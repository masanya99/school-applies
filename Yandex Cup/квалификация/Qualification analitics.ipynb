```python
cards_arr = [i for i in range(2,15)]
cards_arr
```

```python
import random
```

```python
all_cards = cards_arr*4
```

```python
def nCr(n, r): 
  
    return (fact(n) / (fact(r)  
                * fact(n - r))) 
  
# Returns factorial of n 
def fact(n): 
  
    res = 1
      
    for i in range(2, n+1): 
        res = res * i 
          
    return res 
```

```python
# get 6 cards
cards_6 =nCr(len(all_cards), 6)
cards_6
```

```python
def one_time():
    shuffled = random.sample(all_cards, len(all_cards))
    list_of6 = []
    big_arr = []
    count_sums = 0
    for i in shuffled:
        list_of6.append(i)
        if len(list_of6) == 6:
            big_arr.append((list_of6, sum(list_of6)))
            if sum(list_of6) == 21:
                count_sums+=1
            list_of6 = []
        
    return big_arr,count_sums
```

```python
one_time()[1]
```

```python
from collections import Counter
```

```python
def many_tries(N = 20358520, n = 100000):
    tries = np.zeros(n)
    for i in range(n):
        tries[i] = one_time()[1]
    #     tries.append(one_time()[1])
    a = tries.sum()
    return round(a/N, 6)
```

```python
num = 50
diff_answers = np.zeros(num)
for i in range(num):
    diff_answers[i] = many_tries()
diff_answers
```
