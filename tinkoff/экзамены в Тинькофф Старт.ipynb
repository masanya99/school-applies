```python
def time_floor(n,t,floors, pers):
    if t>floors[pers-1]:
        return max(floors)-1
    else: 
        return max(floors[:pers-1]+floors[pers:])+floors[pers-1]-2
   
print(time_floor(n,t,floors, pers))
```
